```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)"

reg2 = requests.get(url).text
```

```python
soup = BeautifulSoup(reg2, "html.parser")
```

```python
rows_label = soup.find_all("table")[0]
rows_label
```

```python
rows = rows_label.find_all("th")
rows_items = [row.text.strip() for row in rows][:6]
```

```python
rows_items
```

```python
df = pd.DataFrame(columns = rows_items)
df
```

```python
labels = rows_label.find("tbody")
labels_items = [label.text for label in labels][2:]
labels_item = [labels.strip("\n\xa0\n") for labels in labels_items]
labels_item2 = [label2.split("\n") for label2 in labels_item if label2 != '']
```

```python
labels_item2[3]
```

```python
for x in labels_item2:
    length = len(df)
    df.loc[length] = x
df
```

```python
df.to_csv("POPULATION_BY_COUNTRIES.csv")
```
